```python
import re

# get all the text from file text_scrape.py
with open('scraped_text.txt', 'r') as file:
    text = file.read()

# Define a regex pattern to match the artist name, album name, release year, extra details, track name, rating and number of ratings
pattern = r"(?P<artist>.+?)\n(?P<album>.+?) \((?P<year>\d{4})\) ?(?P<extra>\[.+?\])?\n\n(?P<track>.+?) - (?P<rating>\d\.\d{2}) \((?P<num_ratings>\d+) ratings\)"


# Create an empty dictionary to store the details
details = {}

# Loop through each match in the text
for match in re.finditer(pattern, text):
    # Get the track name from the match group
    track = match.group("track")
    # Create a sub-dictionary for the track if it doesn't exist
    if track not in details:
        details[track] = {}
    # Get the artist name from the match group
    artist = match.group("artist")
    # Create a sub-dictionary for the artist if it doesn't exist
    if artist not in details[track]:
        details[track][artist] = {}
    # Get the album name from the match group
    album = match.group("album")
    # Create a sub-dictionary for the album if it doesn't exist
    if album not in details[track][artist]:
        details[track][artist][album] = {}
    # Get the release year from the match group
    year = match.group("year")
    # Store the year as a key-value pair in the album sub-dictionary
    details[track][artist][album]["year"] = year
    # Get the extra details from the match group
    extra = match.group("extra")
    # Store the extra details as a key-value pair in the album sub-dictionary if they exist
    if extra:
        details[track][artist][album]["extra"] = extra
    # Get the rating and number of ratings from the match group
    rating = match.group("rating")
    num_ratings = match.group("num_ratings")
    # Store them as key-value pairs in the track sub-dictionary
    details[track][artist][album]["rating"] = rating
    details[track][artist][album]["num_ratings"] = num_ratings

# Print the dictionary
print(len(details))

# Print the dictionary
import pprint
pprint.pprint(details)
```

```python
# Import spacy and load a language model
import spacy
nlp = spacy.load("en_core_web_sm")

# Read the text file and split it into lines
with open("scraped_text.txt", "r") as file:
    text = file.read()
lines = text.split("\n")

# Define patterns for matching track names and artist/album pairs
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
track_pattern = [{"SHAPE": "Xxxx"}, {"TEXT": "-"}, {"SHAPE": "d.dd"}, {"TEXT": "("}, {"IS_DIGIT": True}, {"TEXT": "ratings)"}]
artist_album_pattern = [{"IS_TITLE": True}, {"TEXT": "\n"}, {"IS_TITLE": True}, {"TEXT": "("}, {"IS_DIGIT": True}, {"TEXT": ")"}, {"OP": "?"}, {"TEXT": "\n"}]
matcher.add("TRACK", [track_pattern])
matcher.add("ARTIST_ALBUM", [artist_album_pattern])

# Create empty dictionaries to store track names and artist/album pairs
tracks = {}
artists_albums = {}

# Loop through each line and check if it matches a pattern
for line in lines:
    # Process the line with spacy
    doc = nlp(line)
    # Find matches in the line
    matches = matcher(doc)
    # If there is a match
    if matches:
        # Get the match ID and span
        match_id, start, end = matches[0]
        span = doc[start:end]
        # If the match is a track name
        if nlp.vocab.strings[match_id] == "TRACK":
            # Extract the track name, rating, and number of ratings
            track_name = span[0].text
            rating = span[2].text
            num_ratings = span[4].text
            # Store them in the tracks dictionary
            tracks[track_name] = [rating, num_ratings]
        # If the match is an artist/album pair
        elif nlp.vocab.strings[match_id] == "ARTIST_ALBUM":
            # Extract the artist name, album name, release year, and extra detail
            artist_name = span[0].text
            album_name = span[2].text
            release_year = span[4].text
            extra_detail = span[6].text if len(span) > 6 else None
            # Store them in the artists_albums dictionary
            artists_albums[artist_name] = [album_name, release_year, extra_detail]

# Create a final dictionary that combines tracks and artists_albums
details = {}

# Loop through each track name in tracks
for track_name in tracks:
    # Find the most similar artist name in artists_albums using similarity method
    track_doc = nlp(track_name)
    max_sim = 0
    best_artist = None
    for artist_name in artists_albums:
        artist_doc = nlp(artist_name)
        sim = track_doc.similarity(artist_doc)
        if sim > max_sim:
            max_sim = sim
            best_artist = artist_name
    # Get the album name from the artists_albums dictionary
    album_name = artists_albums[best_artist][0]
    # Create a sub-dictionary for the track name if it doesn't exist
    if track_name not in details:
        details[track_name] = {}
    # Create a sub-dictionary for the artist name if it doesn't exist
    if best_artist not in details[track_name]:
        details[track_name][best_artist] = {}
    # Create a sub-dictionary for the album name if it doesn't exist
    if album_name not in details[track_name][best_artist]:
        details[track_name][best_artist][album_name] = {}
    # Add the release year, extra detail, rating, and number of ratings to the album sub-dictionary
    details[track_name][best_artist][album_name]["year"] = artists_albums[best_artist][1]
    details[track_name][best_artist][album_name]["extra"] = artists_albums[best_artist][2]
    details[track_name][best_artist][album_name]["rating"] = tracks[track_name][0]
    details[track_name][best_artist][album_name]["num_ratings"] = tracks[track_name][1]

# Print the final dictionary
import pprint
pprint.pprint(details)
```

```python
# pprint.pprint(stuff[0])

def search_track_info(dic):
    track_list = []
    for artist in dic:
        for album in dic[artist]:
            year = dic[artist][album]["year"]
            for track in dic[artist][album]:
                if track not in ["year", "extra"]:
                    rating = dic[artist][album][track]["rating"]
                    num_ratings = dic[artist][album][track]["num_ratings"]
                    query = f"artist:{artist} album:{album} track:{track}"
                    results = sp.search(query, type="track", limit=1)
                    if results["tracks"]["total"] > 0:
                        track_id = results["tracks"]["items"][0]["id"]
                        track_info = sp.track(track_id)
                        # create a new dict for the track
                        track_dict = {}
                        # add everything except available markets
                        for key, value in track_info.items():
                            # skip the key that is not useful
                            if key == "available_markets":
                                continue
                            # add the key and value to the dict
                            track_dict[key] = value
                        # add the year, rating and number of ratings from the original dict
                        track_dict["year"] = year
                        track_dict["rating"] = rating
                        track_dict["num_ratings"] = num_ratings
                        # append the dict to the list
                        track_list.append(track_dict)
                        print(f"Found and stored track: {track_dict['name']} by {track_dict['artists'][0]['name']} from {track_dict['album']['name']} ({year})")
                        print()
                    else:
                        print(f"No track found for query: {query}")
                        print()
    return track_list

search_track_info(stuff)
```
